```python
import requests
import pandas as pd
from config import api_key
import json
```

# 1- Park Alerts

```python
#Set the URL of the NPS alerts API and set any required parameters
url = "https://developer.nps.gov/api/v1/alerts"
params = {
    "api_key": api_key,
    #"limit": "50"
}
```

```python
alerts = []

# Retrieve alerts in batches of 50 using pagination
while True:
    response = requests.get(url, params=params)
    data = response.json()  
    alert_data = data['data']  
    alerts.extend(alert_data)
    
    if len(alerts) >= int(data['total']):
        break
    
    params['start'] = str(len(alerts))  # Set the start parameter to retrieve the next batch of alerts

alerts_df = pd.DataFrame(alerts)

# Preview the DataFrame
alerts_df.head()
```

# 2- Park 

```python
url_parks = "https://developer.nps.gov/api/v1/parks"
params = {
    'api_key': api_key
#     'limit': '50'
}
```

```python
parks = []
while True:
    response = requests.get(url_parks, params=params)
    data = response.json()
    park_data = data['data']
    parks.extend(park_data)
    if len(parks) >= int(data['total']):
        break
    params['start'] = str(len(parks))  # Set the start parameter to retrieve the next batch of parks
parks_df = pd.DataFrame(parks)
parks_df.head()    
```

# 3- Park visitor center 

```python
url_visitor = "https://developer.nps.gov/api/v1/visitorcenters"
params = {
    'api_key': api_key
#     'limit': '50'
}
```

```python
visitor_centers = []
while True:
    response = requests.get(url_visitor, params=params)
    data = response.json()
    visitor_data = data['data']
    visitor_centers.extend(visitor_data)
    if len(visitor_centers) >= int(data['total']):
        break
    params['start'] = str(len(visitor_centers))  # Set the start parameter to retrieve the next batch of parks
```

```python
visitor_centers_df = pd.DataFrame(visitor_centers)
visitor_centers_df.head()
```

# Park Events  

```python
url_events = "https://developer.nps.gov/api/v1/events"
params = {
    'api_key': api_key
#     'limit': '50'
}
```

```python
events = []
while True:
    response = requests.get(url_events, params=params)
    data = response.json()
    event_data = data['data']
    events.extend(event_data)
    if len(events) >= int(data['total']):
        break
    params['start'] = str(len(events))  # Set the start parameter to retrieve the next batch of parks
```

```python
events_df = pd.DataFrame(events)
events_df.head()
```

```python
# Remove unecessary columns from the DataFrame and save the new DataFrame 
cleaned_alerts_df = alerts_df[['parkCode','title','category','description','url']]
cleaned_parks_df = parks_df[['fullName','parkCode','description','states','latitude','longitude']]
```

# Used 3 parks (Alerta,Parks,and Visitor center)

```python
# Remove unecessary columns from the DataFrame and save the new DataFrame 
cleaned_alerts_df = alerts_df[['parkCode','title','category','description','url']]
cleaned_parks_df = parks_df[['fullName','parkCode','states','latitude','longitude']]
cleand_visitor_centers_df=visitor_centers_df[['parkCode','directionsInfo','contacts','operatingHours','addresses']]
```

# Merge 

```python
# merge cleaned_alerts_df and cleaned_parks_df on parkCode
merged_df = pd.merge(cleaned_alerts_df, cleaned_parks_df, on='parkCode', how='inner')

# merge merged_df and cleand_visitor_centers_df on parkCode
merged_df = pd.merge(merged_df, cleand_visitor_centers_df, on='parkCode', how='inner')
merged_df.head()
```

```python
# Define the column order you want
column_order = ['parkCode', 'fullName','states', 'title', 'category', 'description',
                'latitude', 'longitude','directionsInfo',
                'contacts', 'operatingHours', 'addresses','url']

# Create a new DataFrame with the columns in the desired order
merged_df = merged_df[column_order]
```

```python
# Define a dictionary with the new column names
new_column_names = {
    'parkCode': 'Park Code',
    'fullName': 'Full Name',
    'title': 'Title',
    'category': 'Category',
    'description': 'Description',
    'url': 'URL',
    'states': 'States',
    'latitude': 'Latitude',
    'longitude': 'Longitude',
    'directionsInfo': 'Directions Info',
    'contacts': 'Contacts',
    'operatingHours': 'Operating Hours',
    'addresses': 'Addresses'
}

# Rename the columns of the merged_df DataFrame
merged_df = merged_df.rename(columns=new_column_names)
```
